```python
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
%matplotlib inline 
```

```python
Whales_path = Path("Resources/whale_returns.csv")
Whales_data = pd.read_csv(Whales_path, parse_dates=True, index_col='Date', infer_datetime_format=True)
Whales_data.head()
```

```python
Whales_data.isnull().sum()
```

```python
Whales_data.dropna(inplace=True)
```

```python
Whales_data
```

```python
Whales_data
```

```python
algo_returns_path = Path("Resources/algo_returns.csv")
algo_returns_data = pd.read_csv(algo_returns_path, parse_dates=True, index_col='Date', infer_datetime_format=True)
algo_returns_data.head()
```

```python
algo_returns_data.isnull().sum()
```

```python
algo_returns_data.dropna(inplace=True)
```

```python
algo_returns_data
```

```python
sp500_history_path = Path("Resources/sp500_history.csv")
sp500_history_data = pd.read_csv(sp500_history_path, parse_dates=True, index_col='Date', infer_datetime_format=True)
sp500_history_data.head()
```

```python
sp500_history_data.isnull().sum()
```

```python
sp500_history_data.dtypes
```

```python
sp500_history_data["Close"] = sp500_history_data["Close"].str.replace("$", "")
sp500_history_data["Close"]=sp500_history_data["Close"].astype(float)
```

```python
sp500_returns = sp500_history_data.pct_change().dropna()
sp500_returns.head()
```

```python
combine_data_returns = pd.concat([Whales_data, algo_returns_data, sp500_returns], axis="columns", join="inner")
combine_data_returns
```

```python
combine_data_returns.head()
```

```python
combine_data_returns.rolling(window=60).std().plot()
```

```python
combine_data_returns.rolling(window=21).std().plot(figsize=(20, 10), title="21 Day Rolling standard deviation")
```

```python
combine_data_correlations = combine_data_returns.corr()
combine_data_correlations
```

```python
combine_data_correlations = combine_data_returns["SOROS FUND MANAGEMENT LLC"].cov(combine_data_returns['Close'])
```

```python
combine_data_correlations
```

```python
sharpe_ratio = (combine_data_returns.mean() * 252) / (combine_data_returns.std() * np.sqrt(252))
```

```python
sharpe_ratio
```
